```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
import numpy as np
import pandas as pd

def evaluate_models(models, X, y, cv=5):
    """
    ใช้ K-Fold Cross Validation เพื่อประเมินหลายโมเดล และบันทึกผลลัพธ์ลง DataFrame

    Parameters:
        models (dict): Dictionary ของโมเดล {ชื่อโมเดล: โมเดล}
        X (array-like): Features
        y (array-like): Target
        cv (int): จำนวนของ Fold (ค่า default = 5)

    Returns:
        results_df (DataFrame): DataFrame ที่เก็บค่า Accuracy, Precision, Recall, และ F1 Score
    """
    results = []

    for name, model in models.items():
        kf = KFold(n_splits=cv, shuffle=True, random_state=42)

        # คำนวณคะแนนแต่ละ metric
        acc_scores = cross_val_score(model, X, y, cv=kf, scoring='accuracy')
        prec_scores = cross_val_score(model, X, y, cv=kf, scoring=make_scorer(precision_score, average='weighted', zero_division=0))
        rec_scores = cross_val_score(model, X, y, cv=kf, scoring=make_scorer(recall_score, average='weighted', zero_division=0))
        f1_scores = cross_val_score(model, X, y, cv=kf, scoring=make_scorer(f1_score, average='weighted', zero_division=0))

        # บันทึกผลลัพธ์
        results.append({
            'Algorithm': name,
            'Accuracy Score': np.mean(acc_scores),
            'Precision Score': np.mean(prec_scores),
            'Recall Score': np.mean(rec_scores),
            'F1 Score': np.mean(f1_scores)
        })

    # แปลงเป็น DataFrame
    results_df = pd.DataFrame(results)

    return results_df
```

#Create Model RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# สร้างโมเดล RandomForestClassifier
rf = RandomForestClassifier(random_state=42)

# กำหนดพารามิเตอร์ที่ต้องการค้นหา
param_grid = {
    'n_estimators': [100, 200, 300],          # จำนวนต้นไม้
    'max_depth': [None, 10, 20, 30],          # ความลึกสูงสุดของต้นไม้
    'min_samples_split': [2, 5, 10],          # จำนวนตัวอย่างขั้นต่ำในการแยก
    'min_samples_leaf': [1, 2, 4],            # จำนวนตัวอย่างขั้นต่ำในใบ
    'max_features': ['sqrt', 'log2', None]    # จำนวนฟีเจอร์ที่ใช้ในการแบ่ง
}

# สร้าง KFold สำหรับ GridSearchCV
cv_kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# สร้าง GridSearchCV ด้วย KFold
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=cv_kfold,                # ใช้ KFold ที่กำหนด
    n_jobs=-1,                  # ใช้ CPU ทั้งหมดที่มี
    verbose=2,                  # แสดงข้อมูลในระหว่างการทำงาน
    scoring='accuracy'          # ใช้ accuracy เป็นเกณฑ์
)

# เทรนโมเดล 10 ครั้งด้วย KFold และเก็บผลลัพธ์
def train_multiple_times_with_kfold(model, X, y, n_iterations=10, n_splits=10):
    results = []

    for i in range(n_iterations):
        print(f"Training iteration {i+1}/{n_iterations}")

        # สร้าง KFold สำหรับการแบ่งข้อมูลในแต่ละรอบ (เพื่อให้การแบ่งข้อมูลแต่ละรอบแตกต่างกัน)
        kf = KFold(n_splits=n_splits, shuffle=True, random_state=42+i)

        # สร้าง clone ของโมเดลเพื่อให้แต่ละการเทรนเป็นอิสระจากกัน
        from sklearn.base import clone
        current_model = clone(model)

        fold_scores = []
        fold_models = []

        # ทำ K-Fold แต่ละรอบ
        for fold, (train_idx, test_idx) in enumerate(kf.split(X)):
            X_train, X_test = X[train_idx], X[test_idx]
            y_train, y_test = y[train_idx], y[test_idx]

            # เทรนโมเดล
            fold_model = clone(current_model)
            fold_model.fit(X_train, y_train)

            # ทำนายผลลัพธ์
            y_pred = fold_model.predict(X_test)

            # คำนวณ accuracy
            acc = accuracy_score(y_test, y_pred)
            fold_scores.append(acc)
            fold_models.append(fold_model)

            print(f"  Fold {fold+1}/{n_splits}, Accuracy: {acc:.4f}")

        # หา fold ที่ดีที่สุดในการเทรนแต่ละรอบ
        best_fold = np.argmax(fold_scores)
        best_score = fold_scores[best_fold]
        best_fold_model = fold_models[best_fold]

        # เก็บผลลัพธ์
        results.append({
            'iteration': i+1,
            'fold_accuracies': fold_scores,
            'mean_accuracy': np.mean(fold_scores),
            'best_fold': best_fold + 1,
            'best_fold_accuracy': best_score,
            'model': best_fold_model
        })

        print(f"  Iteration {i+1} completed. Mean Accuracy: {np.mean(fold_scores):.4f}, Best Fold: {best_fold+1}, Best Accuracy: {best_score:.4f}")

    return results

# ฟังก์ชันหลักที่รวมทั้งหมด
def train_best_random_forest_kfold(X, y, param_grid=None, n_iterations=10, n_splits=10):
    # แปลงข้อมูลเป็น numpy array ถ้ายังไม่ได้แปลง (เพื่อให้แน่ใจว่าใช้ KFold ได้)
    X = np.array(X)
    y = np.array(y)

    # ถ้าไม่มีการกำหนด param_grid ให้ใช้ค่าเริ่มต้น
    if param_grid is None:
        param_grid = {
            'n_estimators': [100, 200, 300],
            'max_depth': [None, 10, 20, 30],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4],
            'max_features': ['sqrt', 'log2', None]
        }

    # สร้างโมเดลและ GridSearchCV ด้วย KFold
    rf = RandomForestClassifier(random_state=42)
    cv_kfold = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    grid_search = GridSearchCV(rf, param_grid, cv=cv_kfold, n_jobs=-1, verbose=1, scoring='accuracy')

    # ค้นหาพารามิเตอร์ที่ดีที่สุด
    print("Finding best parameters using GridSearchCV with KFold...")
    grid_search.fit(X, y)

    print("Best Parameters:", grid_search.best_params_)
    print("Best Cross-validation Score:", grid_search.best_score_)

    # สร้างโมเดลด้วยพารามิเตอร์ที่ดีที่สุด
    best_rf = grid_search.best_estimator_

    # เทรนโมเดล n_iterations ครั้งด้วย KFold
    print(f"Training model {n_iterations} times using KFold with {n_splits} splits...")
    training_results = train_multiple_times_with_kfold(best_rf, X, y, n_iterations, n_splits)

    # คำนวณสถิติจากผลการเทรนทั้งหมด
    mean_accuracies = [result['mean_accuracy'] for result in training_results]
    best_fold_accuracies = [result['best_fold_accuracy'] for result in training_results]

    overall_mean = np.mean(mean_accuracies)
    overall_std = np.std(mean_accuracies)

    print(f"\nResults after {n_iterations} iterations with {n_splits}-fold cross-validation:")
    print(f"Overall Mean Accuracy: {overall_mean:.4f} ± {overall_std:.4f}")

    # หาการเทรนที่ดีที่สุดจากทั้งหมด
    best_iteration = np.argmax(best_fold_accuracies)
    best_model = training_results[best_iteration]['model']

    print(f"Best Model from Iteration: {best_iteration+1}, Fold: {training_results[best_iteration]['best_fold']}")
    print(f"Best Accuracy: {best_fold_accuracies[best_iteration]:.4f}")

    # แสดง feature importance ของโมเดลที่ดีที่สุด
    if hasattr(best_model, 'feature_importances_'):
        importances = best_model.feature_importances_
        indices = np.argsort(importances)[::-1]

        plt.figure(figsize=(10, 6))
        plt.title('Feature Importance')
        plt.bar(range(min(10, len(importances))), importances[indices[:10]], align='center')
        plt.xticks(range(min(10, len(importances))), indices[:10])
        plt.tight_layout()

    # สรุปผลการเทรนแต่ละรอบ
    summary_df = pd.DataFrame({
        'Iteration': [r['iteration'] for r in training_results],
        'Mean Accuracy': [r['mean_accuracy'] for r in training_results],
        'Best Fold': [r['best_fold'] for r in training_results],
        'Best Fold Accuracy': [r['best_fold_accuracy'] for r in training_results]
    })

    return best_model, grid_search, training_results, summary_df
```
